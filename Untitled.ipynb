```python
import pandas as pd 
import numpy as np

data = {'open' : [100,101,102], 'high' : [200,201,202], 'low' : [300,301,302], 'close' : [400,401,402]}

df = pd.DataFrame(data)
df.head()
```

```python
print(df)
```

```python
df.shape
```

```python
df.columns
```

```python
df.index = pd.date_range(start='2003-01-24', periods=len(df))
df
```

```python
df['daily change'] = df['close'] - df['open']
df
```

```python
df['precente change'] = df['daily change']*100/df['open']
df
```

```python
df[df['precente change']<300]
```

```python
df['cumulative'] = df['daily change'].cumsum()
df
```

```python
df['pnl'] = [1000,2000,3000]
df
```

```python
df['total pnl'] = df['pnl'].cumsum()
df
```

```python
df['bullish'] = df['close'] > df['open']
df
```

```python
df['signal'] = np.where(df['close'] > 401, 'buy', 'nan')
df
```
